<h4 style="background:cyan;text-align:center;padding:10px">Boucles ```while```</h4>

<code>**if** &lt;condition&gt;:

````python
for film in films:
    if film[1] == "Guy A. Lepage":
        print(film)
````

Mais √ßa ne donne aucun r√©sultat. C'est que M. Lepage a sign√© un film avec un autre r√©alisateur. Ainsi, il n'arrive jamais que le 2e √©l√©ment de la variable ```film``` contienne tr√®s exactement et uniquement ¬´Guy A. Lepage¬ª. Le code suivant est donc pr√©f√©rable:

````python
for film in films:
    if "Guy A. Lepage" in film[1]:
        print(film)
````

Je vous ai aussi montr√© comment trier les films r√©alis√©s en 2000 ou apr√®s. Il s'agit de se servir du 3e √©l√©ment de la variable ```film``` (```film[2]```) qui contient l'ann√©e o√π chaque film a √©t√© r√©alis√©. Cette variable est de type texte. Ainsi, on peut en extraire les deux premiers caract√®res, comme on l'a fait ci-dessus avec du texte. Il suffit d'ajouter ```[:2]``` √† ```film[2]```.

Les films contenus dans la liste ont √©t√© r√©alis√©s entre 1943 et 2020. Les deux premiers caract√®res de ces ann√©es ne peuvent √™tre que ¬´19¬ª ou ¬´20¬ª. Ainsi, il suffit de faire la condition ci-dessous pour ne faire afficher que les films r√©alis√©s en 2000 ou apr√®s:

````python
for film in films:
    if film[2][:2] == "20":
        print("Le film {} a √©t√© r√©alis√© en {}.".format(film[0],film[2]))
````

Je vous ai ensuite montr√© une technique semblable pour trier les films r√©alis√©s avant les ann√©es 1960. Essayons d'appliquer la m√™me chose, mais aux films r√©alis√©s durant les ann√©es 1980 et 1990.

Cela implique l'utilisation de la fonction ```int()``` pour transformer en nombres les trois premiers caract√®res des ann√©es.

````python
for film in films:
    if int(film[2][:3]) > 197 and int(film[2][:3]) <= 199:
        print("Le film {} a √©t√© r√©alis√© en {}.".format(film[0],film[2]))
````
